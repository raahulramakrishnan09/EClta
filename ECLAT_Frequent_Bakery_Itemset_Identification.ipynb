```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/bread basket.csv')
df=df.iloc[:,[0,1]]
df
```

```python
#grouping
grp=df.groupby('Transaction')['Item'].apply(list).reset_index()
grp
```

```python
#transforming_to_panda_series
df=grp['Item'].apply(pd.Series)
df
```

```python
#installing_pyECLAT
!pip install pyECLAT
```

```python
#ECLAT_Algorithm
from pyECLAT import ECLAT
eclat_instance=ECLAT(data=df,verbose=True)
```

```python
#unique_items_in_the_dataset
eclat_instance.uniq_
```

```python
#specifying_parameters
get_eclat_indexes,get_eclat_supports=eclat_instance.fit(min_support=0.01,min_combination=2,verbose=True,separator=' & ')
```
